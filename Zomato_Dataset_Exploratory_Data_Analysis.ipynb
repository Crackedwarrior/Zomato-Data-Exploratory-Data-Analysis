# **Zomato Dataset Exploratory Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv',encoding = 'latin-1')
df.head()
```

```python
df.columns
```

```python
df.describe()
```

# **Data Analysis Objectives**

1. **Handling Missing Data**
   - Identification and imputation of missing values

2. **Numerical Variable Exploration**
   - Descriptive statistics and distribution analysis of numerical features

3. **Feature Relationship Analysis**
   - Correlation analysis and identification of relationships between features

4. **Categorical Variable Exploration**
   - Analysis and visualization of categorical feature distributions


```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels= False, cbar = False, cmap = 'viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code', how ='left')
final_df
```

```python
#Data Types of the attributes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names =final_df.Country.value_counts().index
```

```python
country_value = final_df.Country.value_counts().values
```

```python
#Pie chart representing Active usage of Zomato App
plt.pie(country_value,labels = country_names)
```

```python
#pie chart of top 3 countries
plt.pie(country_value[:3],labels = country_names[:3],autopct= "%1.2f%%")
```

## ***Numerical Variables***

```python
final_df.columns
```

```python
ratings = final_df.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating", y = "Rating Count",data = ratings)
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating Count",data = ratings,hue = 'Rating color',palette = ['blue','red','orange','yellow','green','darkgreen'])
```

```python
##Count Plot
sns.countplot(x = "Rating color",data = ratings,hue = 'Rating color',palette = ['blue','red','orange','yellow','green','darkgreen'])
```

```python
final_df.City.value_counts()
city_values = final_df.City.value_counts().values
city_names = final_df.City.value_counts().index

plt.pie(city_values[:5],labels = city_names[:5],autopct= "%1.2f%%")
```

```python
#Top 5 cuisines
final_df.Cuisines.value_counts()
cuisine_values = final_df.Cuisines.value_counts().values
cuisine_names = final_df.Cuisines.value_counts().index

plt.pie(city_values[:5],labels = city_names[:5],autopct= "%1.2f%%")
```
